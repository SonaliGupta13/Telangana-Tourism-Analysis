```python
#glob function for merging multiple files into one
import pandas as pd
from glob import glob
```

```python
#show all the files that has to be merged that's why using * after the same file name
foreign_visitors=sorted(glob('E:/2022/Power BI/Challanges/codebasics/challenge5/C5 Input for participants/C5 Input for participants/foreign_visitors/foreign_visitors*.csv'))
foreign_visitors
```

```python
#expression to read all the files iteratively and create a new column to identify the record original file name using assign()
mergeddata=pd.concat(pd.read_csv(foreign_visitor).assign(sourcefilename=foreign_visitor)
for foreign_visitor in foreign_visitors)
mergeddata    
```

```python
#save updated merge file into the given path
mergeddata.to_csv('E:/2022/Power BI/Challanges/codebasics/challenge5/C5 Input for participants/C5 Input for participants/foreign_visitors/foreign_visitors.csv')
```
